# 머신러닝을 어떻게 정의?
- 머신러닝이란 데이터로부터 학습할 수 있는 시스템을 만드는 것. 이 때 학습이란 어떤 작업에서 주어진 성능 지표가 더 나아지는 것을 의미한다

# 머신러닝이 도움을 줄 수 있는 문제 유형?
- 명확한 해결책이 없는 복잡한 문제, 수작업으로 된 긴 규칙 리스트를 대체하는 경우, 변화하는 환경에 적응하는 시스템을 만드는 경우, 사람에게 통찰을 제공해야 하는 경우(ex. 데이터 마이닝)에 머신러닝이 도움을 줄 수 있다

# 레이블 된 훈련 세트란?
- 레이블된 훈련 세트란 각 샘플에 대해 원하는 정답(레이블)을 담고 있는 훈련 세트를 의미한다

# 가장 널리 사용되는 지도 학습 작업 두 가지?
- 가장 일반적인 두 가지 지도 학습 문제는 분류와 회귀이다

# 보편적인 비지도 학습 작업 네 가지?
- 보편적인 비지도 학습 문제는 군집, 시각화, 차원 축소, 연관 규칙 학습이다

# 사전 정보가 없는 여러 지형에서 로봇을 걸어가게 하려면 어떤 종류의 머신러닝 알고리즘을 사용할 수 있는가?
- 알려지지 않은 지형을 탐험해야 하는 경우 가장 좋은 방법은 강화 학습이다
- 이는 전형적으로 강화 학습이 다루는 유형의 문제이며, 지도 학습이나 비지도 학습으로 표현할 수도 있지만 일반적이지 않다

# 고객을 여러 그룹으로 분할하려면 어떤 알고리즘을 사용해야 하는가?
- 그룹을 어떻게 분할해야 할지 모르는 상황에서는 비슷한 고객끼리의 군집으로 나누기 위해 비지도 학습의 군집 알고리즘을 사용
- 어떤 그룹으로 분할이 되어야 할 지 안다면 지도 학습의 분류 알고리즘을 통해 각 그룹에 대한 샘플을 주입

# 스팸 감지의 문제는 지도 학습과 비지도 학습 중 어떤 문제로 볼 수 있는가?
- 스팸 감지는 많은 이메일과 이에 상응하는 레이블(스팸 혹은 스팸이 아님)이 주어지는 전형적인 지도 학습 문제이다

# 온라인 학습 시스템이란?
- 온라인 학습 시스템은 배치 학습 시스템과 달리 점진적으로 학습할 수 있다
- 이 방식은 변화하는 데이터와 자율 시스템에 빠르게 적응하고 매우 많은 양의 데이터를 훈련시킬 수 있다

# 외부 메모리 학습이란?
- 외부 메모리 알고리즘은 컴퓨터의 주메모리에 들어갈 수 없는 대용량의 데이터를 다룰 수 있다
- 외부 메모리 학습 알고리즘은 데이터를 미니배치로 나누고 온라인 학습 기법을 사용해 학습한다

# 예측을 하기 위해 유사도 측정에 의존하는 학습 알고리즘은 무엇인가?
- 사례 기반 학습 시스템은 훈련 데이터를 기억하는 학습이다
- 새로운 샘플이 주어지면 유사도 측정을 사용해 학습된 샘플 중에서 가장 비슷한 것을 찾아 예측으로 사용한다

# 모델 파라미터와 학습 알고리즘의 하이퍼파라미터 사이에는 어떤 차이가 있는가?
- 모델은 하나 이상의 파라미터(ex.선형 모델의 기울기)를 사용해 새로운 샘플이 주어지면 무엇을 예측할지 결정
- 학습 알고리즘은 모델이 새로운 샘플에 잘 일반화 되도록 이런 파라미터들의 최적값을 찾는다
- 하이퍼파라미터는 모델이 아니라 이런 학습 알고리즘 자체의 파라미터이다(ex.적용할 규제의 정도)

# 모델 기반 알고리즘이 찾는 것은? 성공을 위해 이 알고리즘이 사용하는 가장 일반적인 전략은 ? 예측은 어떻게?
- 모델 기반 알고리즘은 새로운 샘플에 잘 일반화되기 위한 모델 파라미터의 최적값을 찾는다
- 일반적으로 훈련 데이터에서 시스템의 예측이 얼마나 나쁜지 측정하고 모델에 규제가 있다면 모델 복잡도에 대한 패널티를 더한 비용 함수를 최소화함으로써 시스템을 훈련
- 예측을 만들려면 학습 알고리즘이 찾은 파라미터를 사용하는 모델의 예측 함수에 새로운 샘플의 특성을 주입

# 머신러닝의 주요 도전 과제는 무엇인가?
- 부족한 데이터, 낮은 데이터 품질, 대표성 없는 데이터, 무의미한 특성, 훈련 데이터에 과소적합된 지나치게 간단한 모델, 훈련 모델 데이터에 과대적합된 지나치게 복잡한 모델 등이다

# 모델이 훈련 데이터에서의 성능은 좋지만 샘플에서의 일반화 성능이 나쁘다면 어떤 문제가 있는 것인가? 이 때 가능한 해결책은 ?
- 이 경우 훈련 데이터에 과대적합되었을 가능성이 높다. 이에 대한 해결책으로는,
 - 더 많은 데이터 모으기
 - 모델을 단순화하기(간단한 알고리즘을 선택하거나, 특성이나 파라미터의 수를 줄이거나, 모델의 규제를 추가)
 - 훈련 데이터에 있는 노이즈를 감소시키기
- 이와 같은 방법이 있다

# 테스트 세트가 무엇이고, 왜 사용해야 하는가?
- 테스트 세트는 실전에 배치되기 전에 모델이 새로운 샘플에 대해 만들 일반화 오차를 추정하기 위해 사용

# 검증 세트의 목적은 무엇인가?
- 검증 세트는 모델을 비교하는 데 사용된다. 이를 사용해 가장 좋은 모델을 고르고 하이퍼 파라미터를 튜닝한다

# 훈련-개발 세트가 무엇인가 언제 필요하고 어떻게 사용해야 하는가?
- 모델을 실전에 투입했을 때 사용될 데이터와 가능한 최대로 가까워야하는 검증, 테스트 세트에 사용되는 데이터와 훈련 세트 사이에 데이터 불일치 위험이 있을 때 사용한다
- 훈련 세트의 일부에서 모델을 훈련하고 훈련-개발 세트와 검증 세트에서 평가
- 모델이 훈련 세트에서 잘 동작하지만 훈련-검증 세트에서 나쁜 성능을 낸다면 과대적합 되었을 가능성이 높음
- 훈련 세트와 훈련-개발 세트에서 잘 동작하지만 개발 세트에서 성능이 나쁘다면 훈련 데이터와 검증+테스트 데이터 사이에 데이터 불일치가 있을 가능성이 높다
- 검증+테스트 데이터에 더 가깝게 되도록 훈련 데이터 개선해야 한다

# 테스트 세트를 사용해서 하이퍼 파라미터를 튜닝하면 어떤 문제가 생기는가?
- 테스트 세트를 사용해 하이퍼 파라미터를 튜닝하면 테스트 세트에 과대적합 될 위험이 있고 일반화 오차를 낙관적으로 측정하게 된다
- 이렇게 모델을 출시하면 기대한 것보다 나쁜 성능을 낼 것이다
